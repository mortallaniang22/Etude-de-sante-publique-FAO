# I. Importation et nettoyages des données:

```python
dispo_alimentaire = pd.read_csv("dispo_alimentaire.csv")

dispo_alimentaire.head()
```

```python
dispo_alimentaire.shape
```

**Convertir les colonnes pour homogénéiser les unités de mesures** 

g/personne/jour = 0,001kgx365/personne/an

```python
#Convertir les colonnes g/personne/jour en kg/personne/an.
Conversion = ["Disponibilité de matière grasse en quantité (g/personne/jour)", "Disponibilité de protéines en quantité (g/personne/jour)"]
for x in Conversion:
    dispo_alimentaire[x] *= 365*0.001
dispo_alimentaire.head()                                                                        
```

```python
## Convertir en kg les 10 variables qui précisent les utilisations/moyens d’acquisition de la disponibilité alimentaire exprimer en tonnes.
## 1 tonne = 1000 kg alors 1000 tonnes = 1 000 000 Kg

Conversion_utilisation = ["Aliments pour animaux","Disponibilité intérieure","Autres Utilisations",  "Exportations - Quantité","Importations - Quantité",
                          "Nourriture","Pertes","Production", "Semences","Traitement","Variation de stock"]
for x in Conversion_utilisation:
    dispo_alimentaire[x]  *= 1000000
dispo_alimentaire.head()                                                                                  
```

**Renommer les colonnes**

```python
## Renommer les colonnes g/personne/jour en kg/personne/an.
dispo_alimentaire.rename(columns = {"Disponibilité de matière grasse en quantité (g/personne/jour)" : "Disponibilité de matiére grasse en quantité (kg/personne/an)", 
                                    "Disponibilité de protéines en quantité (g/personne/jour)" : "Disponibilité de protéines quantité (kg/personne/an)"}, inplace=True)

dispo_alimentaire.head()
```

**Remplacer les valeurs manquantes par 0**

```python
# Les données manquantes seront remplacées par 0.
dispo_alimentaire.fillna(0, inplace=True)
dispo_alimentaire.head()
```

## 1. LES DEMANDES DE MARC

### A. La proportion de personne en état de sous-nutrition.

Nous allons procerdez à la jointure des dataframes population et sous-nutrtion pour l'année 2017 


```python
# Fusionner les dataframes en utilisant pd.merge et on applique une filtre à la colonne année pour avoir les données de 2017.
population_sous_nutrition = pd.merge(population.loc[population["Année"] == 2017], 
                     sous_nutrition.loc[sous_nutrition["Année"] == 2017],  on = "Zone")
population_sous_nutrition.head()
```

```python
## Renommer les colonnes valeur_x et valeur_y respectivement par Population et sous_nutrition.
population_sous_nutrition.rename(columns = {"Valeur_x" : "Population",  "Valeur_y" : "sous_nutrtion",   
                                         "Année_x" : "Année"}, inplace = True)
population_sous_nutrition.head()                                                                                 
```

```python
#Calculer la population mondiale en utulisant .sum()
population_mondiale = population_sous_nutrition["Population"].sum()
print("La population mondiale est estimée à", population_mondiale,"personnes.")                        
```

```python
##Calculer le nombre de personne en sous-nutrition en utulisant .sum()

personne_en_sous_nutrition = population_sous_nutrition["sous_nutrtion"].sum()
print( "Le nombre de personne en sous-alimentation est de", personne_en_sous_nutrition, "dans le monde.")       
```

```python
#Calculer la proportion de personne en sous-nutrition par rapport à la population mondiale.

print ("La proportion de personne en sous-nutrition est de", 
       round(((personne_en_sous_nutrition)/(population_mondiale))*100,2),"%")                                  
```

### B. Le nombre théorique de personnes qui pourraient etre nourries.

```python
## Fusionner les fichiers disponibilité alimentaire et population.
## Utuliser pd.merge pour la fision et appliquer une filtre à la colonne année pour avoir les données de 2017.
## Indiquer les colonnes que l'ont souhaite afficher apres la fusion.

dispo_population = pd.merge((dispo_alimentaire[["Zone","Disponibilité alimentaire (Kcal/personne/jour)"]]),
                     (population.loc[population["Année"] == 2017, ["Zone","Valeur"]]),
                     on = "Zone")
dispo_population.head()
```

```python
##Renommer la colonne valeur par Population.

dispo_population.rename(columns = {"Valeur" : "Population"}, inplace = True)
dispo_population.head()                                                                                     
```

```python
## Ajouter une colonne calculée nommée disponibilité aliementaire par zone qui est egale au produit dedisponibilité alimentaire et population.

dispo_population["disponibilité aliementaire par zone"] = (dispo_population["Population"] * dispo_population["Disponibilité alimentaire (Kcal/personne/jour)"])    
dispo_population.head()
```

```python
# Créer une variable qui est egale à la somme de la disponibilité alimentaire par zone.

dispo_aliment_mondiale = dispo_population["disponibilité aliementaire par zone"].sum()
print("La disponibilité alimentaire mondiale  est de ", dispo_aliment_mondiale, "kcal/personne/jour.")
print("")

## Diviser la disponibilité alimentaire mondiale par 2500 qui est la calories moyenne journaliére d'une personne.

print("Le nombre theorique de personnes qui pourraient etre nouries est de",(dispo_aliment_mondiale)/2500,"personnes.") 
```

```python
nb_theorique=dispo_aliment_mondiale/2500
print("La disponibilité alimentaire mondiale peux nourir",round((nb_theorique/population_mondiale)*100,2),"%","de la population mondiale")
```

### C. Le nombre théorique de personnes qui pourraient étre nourries avec des produits végétaux.

```python
## Fusionner les fichiers disponibilité alimentaire et population.
## Utuliser la fonction pd.merge et on applique une filtre à la colonne année pour avoir les données de 2017.    
## Indiquer les colonnes que l'ont souhaite afficher apres la fusion.

dispo_vegetale= pd.merge((dispo_alimentaire[["Zone","Disponibilité alimentaire (Kcal/personne/jour)","Origine"]]),
                     (population.loc[population["Année"] == 2017, ["Zone","Valeur"]]),
                     on = "Zone")
dispo_vegetale.head()
```

```python
## Filtrer la colonne origine pour afficher que les produits végétaux.

dispo_vegetale = dispo_vegetale[dispo_vegetale["Origine"] == "vegetale"]                                                              
dispo_vegetale.head()                                                                   
```

```python
## Ajouter une colonne nommé dispo produit vegetal qui est egale a disponibilité*la population du pays.
dispo_vegetale["dispo produit vegetale"] = dispo_vegetale["Disponibilité alimentaire (Kcal/personne/jour)"]*dispo_vegetale["Valeur"]
dispo_vegetale.head()
```

```python
## Calculer la disponibilité des produits d'origines végétales mondiale.

disponibilité_mondiale = dispo_vegetale["dispo produit vegetale"].sum()
print("Disponibilité mondiale des produits d'origine végétale",disponibilité_mondiale,"Kcal/personne/jours")
print("")
print("Les produits d'origine végétaux peuvent nourrir théoriquement", disponibilité_mondiale/2500, "personnes")
print("")
```

```python
nb_pers_vegetaux =  disponibilité_mondiale/2500
print("Les produits d'origine végétale peuvent nourir",round((nb_pers_vegetaux/population_mondiale)*100,2),"%","de la population mondiale")
```
